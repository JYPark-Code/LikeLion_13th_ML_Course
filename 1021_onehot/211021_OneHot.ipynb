### Label , One hot Encoding

```python
### 01. 데이터 준비
import pandas as pd
data = { "eng": ["b", "c", "a", "d"] }
df = pd.DataFrame(data)
print(type(df))
df
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
en_x = LabelEncoder()

df["LabelEncoding"] = en_x.fit_transform(df["eng"])
```

```python
df
```

```python
onehot = OneHotEncoder()
val = df["LabelEncoding"].values.reshape(-1, 1) # OneHotEncoder() # 판다스 -> 원핫 할때 reshape이 필요하다.

y = onehot.fit_transform([df["LabelEncoding"].values] ).toarray() # 값을 변경후, 배열로 만들어준다.
y
```

```python
y = onehot.fit_transform( val ).toarray() # 값을 변경후, 배열로 만들어준다.
y
```

```python
onehot_val = pd.DataFrame(y, dtype=int)
print(onehot_val)

df_new = pd.concat([df, onehot_val], axis=1)
df_new
```

```python
### 01. 데이터 준비
import pandas as pd
data = { "feature1":[2,3,8,4],
         "feature2":[22,32,82,42],
         "target": ["b","c", "a", "d"]
       }
df = pd.DataFrame(data)
df
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
df['lbl_en'] = label_encoder.fit_transform(df['target'])
df
```

```python
train_y = df['lbl_en'].values.reshape(len(df), 1)

onehot_encoder = preprocessing.OneHotEncoder(sparse=False)
train_y_onehot = onehot_encoder.fit_transform(train_y)
print(train_y_onehot)
print(train_y_onehot.shape)
```
