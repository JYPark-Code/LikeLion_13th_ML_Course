```python
doc_ko = open("15_TheExtreme_utf8.txt").read()
doc_ko[1:1000]
```

```python
from konlpy.tag import Okt 
t = Okt()
```

```python
doc_nouns = t.nouns(doc_ko)
print(len(doc_nouns))
doc_nouns
```

```python
# nltk - 자연어 처리 라이브러리
import nltk
```

```python
ko = nltk.Text(doc_nouns, name="분노의 질주")
```

```python
type(ko)
```

```python
print(len(ko.tokens), type(ko.tokens))
print(ko.tokens[:15])
print(ko.vocab().most_common(20))
```

```python
top_50 = ko.vocab().most_common(50)
top_50
```

```python
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
stopwords = ["이","것","점", "영화", "볼"]

new_ko = [ ]
for one_word in ko:
  if one_word not in stopwords:
    new_ko.append(one_word)
```

```python
## nltk 객체 만들어서 넣기
new_ko = nltk.Text(new_ko, name="분노의 질주(수정)")
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

## 텍스트 분포도 확인

```python
plt.figure(figsize=(15,8))
new_ko.dispersion_plot(["자동차","워커","재미","평점"])
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
Car_mask = np.array(Image.open("Draw_car1.png"))

data = new_ko.vocab().most_common(1000)

wc = WordCloud(max_words=500,
               mask = Car_mask,
               font_path = path,
               background_color="white",
               contour_width=3,
               contour_color="steelblue").generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
