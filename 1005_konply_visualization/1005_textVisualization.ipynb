# 불용어 사전 - 파일 만들기

```python
stop_file = open("stopword_dic.txt")

all_lines = stop_file.readlines()
```

```python
stop_words_text = []
for line in all_lines:
  stop_words_text.extend(line.split(","))
stop_words_text
```

```python
### 한글에서는 따로 불용어 사전이 없어, 따로 만들거나 또는 파일로 부터 불러올 수 있다.
stop_words = ['007', '영화', '액션', '시리즈', '더', '제임스', '본드',
              '그', '이', '것', '또', '좀', 
              '다음', '질주', '그냥', 
              '말', '뭐', '애', '나', '듯', '편', '볼', '점', '중', '로']

new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words_text:
    new_ko.append(one_word)
```

```python
### nltk Text 객체 만들기
new_ko = nltk.Text(new_ko, name="007 노타임투다이")
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
masking_image = np.array(Image.open("007_no_time_die.jpg"))
```

```python
data = new_ko.vocab().most_common(1000)
```

```python
wc = WordCloud(background_color='white', 
               max_words=200,
               mask=masking_image,   
               contour_width=3, 
               contour_color="steelblue", 
               font_path=path).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(20,100))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
